```python
import twint
```

```python
data = pd.read_csv('./Test_Entrepreneur_list.csv', sep=';')
data
```

```python
User_Twitter_Name_List = list(data["Twitter_User_Name"])
User_Twitter_Name_List
```

```python
# Dummy User; For reference Purpose
c = twint.Config()
c.Username = "tferriss"
c.Pandas = True

tferriss_df = twint.storage.panda.User_df
```

```python
tferriss_df
```

```python
Column_names = list(tferriss_df.columns)
Column_names
```

```python
# Delete useless element 
Column_names.remove("name")
Column_names.remove("username")
Column_names.remove("avatar")
Column_names.remove("background_image")
Column_names.remove("join_datetime")
```

```python
Column_names
```

```python
for i in range(len(Column_names)):
    data[Column_names[i]] = " "
```

```python
data
```

```python
# Scrap User information
for i in range(len(User_Twitter_Name_List)):
    c = twint.Config()
    c.Username = User_Twitter_Name_List[i]
    c.Pandas = True
    c.Store_csv = True
    c.Output = f"./Test_Entrepreneur_User_Info/Info{i}.csv"
    
    twint.run.Lookup(c)


```

```python
for i in range(len(User_Twitter_Name_List)):
    df = pd.read_csv("./Test_Entrepreneur_User_Info/"+ f"Info{i}.csv")
    for j in range(len(Column_names)):
        column_name = Column_names[j]
        data.loc[i, column_name] = df.loc[0, column_name]
    
```

```python
data
```

```python
data.to_csv('Test_User_Info_Without_Tweets.csv', index=False)
```
